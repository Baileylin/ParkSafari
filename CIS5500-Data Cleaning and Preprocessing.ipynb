# Data Cleaning

## Formatting column names

```python
# Change column names to lower case and substitute space with "_"
species = species.rename(columns=lambda x: x.lower().replace(" ", "_"))
trails = trails.rename(columns=lambda x: x.lower().replace(" ", "_"))
parks = parks.rename(columns=lambda x: x.lower().replace(" ", "_"))
airbnb = airbnb.rename(columns=lambda x: x.lower().replace(" ", "_"))
```

```python
# Change column names that start with "_" in trails
trails = trails.rename(columns = {'_geoloc':'geoloc'})
```

```python
# Change column name of "order" to "order_name" in species dataset
species = species.rename(columns = {'order':'order_name'})
```

## Drop unnamed and unrelated columns

```python
# Change the column name of "area_name" to "park_name" in the trails dataset
trails = trails.rename(columns = {'area_name':'park_name'})
```

```python
# Drop unnamed column in airbnb
airbnb = airbnb.drop(columns = ['unnamed:_0'])
airbnb.columns
```

```python
# Drop unnamed column in species
species = species.drop(columns = ['unnamed:_13'])
species.columns
```

```python
# Drop unit column in trails because all units are supposed to be in meters
trails = trails.drop(columns = ['units'])
trails.columns
```

## Drop unneeded columns

```python
# Drop unneeded columns in trails such as vistor_usage
trails = trails.drop(columns = ['visitor_usage'])
trails.columns
```

## Make park names across all datasets consistent

```python
# Check how many unique national park names are in the park dataset
len(parks['park_name'].unique()) 
```

```python
# Check if the park names in the species dataset and in the park dataset are consistent with each other
if set(species['park_name']) == set(parks['park_name']):
    print("The park names are consistent")
else:
    print("The park names are not consistent")
```

```python
# Check if the park names in the trails dataset and in the park dataset are consistent with each other
if set(trails['park_name']) == set(parks['park_name']):
    print("The park names are consistent")
else:
    print("The park names are not consistent")
```

```python
# Check the number of unique natioanl park names in the trails dataset
len(trails['park_name'].unique())
```

```python
# Find the park name that are not consistent between parks and trails dataset
set1 = set(parks['park_name'])
set2 = set(trails['park_name'])

diff_parks = set1.symmetric_difference(set2)
diff_parks
```

```python
# Make the park names in trails be consistent with the ones in parks
trails['park_name'] = trails['park_name'].replace('Congaree National Park Wilderness', 'Congaree National Park')
trails['park_name'] = trails['park_name'].replace('Denali National Park', 'Denali National Park and Preserve')
trails['park_name'] = trails['park_name'].replace('Katmai National Park', 'Katmai National Park and Preserve')
trails['park_name'] = trails['park_name'].replace('Glacier Bay National Park', 'Glacier Bay National Park and Preserve')
```

```python
# Keep only the park names that are in the park dataset
common_parks = set(parks['park_name']).intersection(set(trails['park_name']))

trails = trails[trails['park_name'].isin(common_parks)]
```

```python
# Again, check the number of unique national park names in the trails dataset
len(trails['park_name'].unique())
```

```python
# Check if all the park name in trails dataset is in parks dataset
if trails['park_name'].isin(parks['park_name']).all():
    print("All park names in trails dataset are in parks dataset")
else:
    print("Some park names in trails dataset are not in parks dataset")
```

```python
# Check if all the park names in the parks dataset are unique
if parks['park_name'].duplicated().any():
    print("There are duplicate park names in the parks dataset")
else:
    print("All park names in the parks dataset are unique")
```

```python
# Check if each park ID matches uniquely to a park name
unique_parks = parks[['park_code', 'park_name']].drop_duplicates()

if len(unique_parks) == len(parks):
    print("Each park ID matches uniquely to a park name")
else:
    print("There are park IDs that match to multiple park names or vice versa")
```

## Add Column to Trail

```python
# Add park_code column to trails
merged_df = pd.merge(trails, parks, on='park_name')
trails= merged_df[['trail_id','park_code','name','park_name','city_name','state_name','country_name','geoloc',
                   'popularity','length','elevation_gain','difficulty_rating','route_type','avg_rating', 'num_reviews','features','activities']]
```

## Parse Geoloc into Longitude and Latitude

```python
# Check the datatype of Geoloc
```

```python
# Convert 'geoloc' column from string to dictionary, and split into separate 'latitude' and 'longitude' columns
trails[['latitude', 'longitude']] = trails['geoloc'].apply(lambda x: pd.Series(eval(x.replace("'", '"'))) if isinstance(x, str) else pd.Series([None, None]))

# Drop the original 'geoloc' column
trails = trails.drop('geoloc', axis=1)
```
